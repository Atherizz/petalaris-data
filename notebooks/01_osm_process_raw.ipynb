```python
import geopandas as gpd
import pandas as pd 
from pathlib import Path

folder = Path("../data/raw/osm")
files = list(folder.glob("*.geojson"))

gdfs = [gpd.read_file(f) for f in files]
gdf = pd.concat(gdfs, ignore_index=True)    


```

```python
gdf["lon"] = gdf.geometry.x
gdf["lat"] = gdf.geometry.y

amenity  = gdf["amenity"]  if "amenity"  in gdf.columns else pd.Series([None]*len(gdf))
leisure  = gdf["leisure"]  if "leisure"  in gdf.columns else pd.Series([None]*len(gdf))
landuse  = gdf["landuse"]  if "landuse"  in gdf.columns else pd.Series([None]*len(gdf))

gdf["category"] = amenity.fillna(leisure).fillna(landuse)

# ── Filter & save
df = gdf[["name", "lon", "lat", "category"]]
df = df.dropna(subset=["name", "lon", "lat", "category"])

output_path = Path("../data/processed/poi_raw_malang.csv")
output_path.parent.mkdir(parents=True, exist_ok=True)

df.to_csv(output_path, index=False)
print(f"✅ Success export data: {output_path}")
```
